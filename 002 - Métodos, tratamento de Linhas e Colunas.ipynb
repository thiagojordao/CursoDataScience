```python
#Podemos renomear as colunas no momento da leitura do arquivo

colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration',
       'actors_list']
file = 'movies.csv'
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns
```

```python
#É muito comum termos arquivos com várias colunas e a operação de renomear uma a uma fica 
#inviável. Podemos então utilizar funções do Python para trabalharmos com as colunas

file = 'movies.csv'
filmes = pd.read_csv(file)
filmes.head()

filmes.columns = filmes.columns.str.replace(' ','_')
filmes.columns
```

## Removendo colunas do Dataframe

```python
#Por boa prática, ao fazer uma alteração eu um Dataframe, o Dataframe alterado deve ser atribuído 
#a outro Dataframe, mantendo a rastreabilidade

filmes01 = filmes.drop(['content_rating'],axis=1) #coluna
filmes01.head()
```

```python
#Removendo mais de 1 coluna

filmes02 = filmes01.drop(['duration','actors_list'],axis=1) #coluna
filmes02.head()
```
