# O objetivo é retirar a vírgula e inverter a posição dos nomes

```python
import re #re = regex expression

match = re.compile(r'"(.*)(,)(\s.*)"') #Cria o padrão: (.*) = tudo / (,) = apenas a virgula / (\s.*) = tudo após o espaço
match.sub(r'"\3 \1"',linhas[1])  #submete o padrão pegando a terceira expressão (\3) e a primeira expressão (\1) apenas para
                                 #quem está entre aspas duplas
```

```python
#Criando uma função que remove a virgula dos nomes e troca a posição

def tratar_nome(linha):
    import re
    match = re.compile(r'"(.*)(,)(\s.*)"')
    return match.sub(r'"\3 \1"',linha)
```

```python
lista_idade = []

for idade in titanic_dados['Age']:
    if idade != -1:
        lista_idade.append(idade)
```

```python
def media(lista):
    total = sum(lista)
    qtd = len(lista)
    media = total/qtd
    
    return media
```

```python
print(media(lista_idade))
```

```python
media_idades = media(lista_idade)
```

```python
for indice,age in enumerate(titanic_dados['Age']):
    if age == -1:
        titanic_dados['Age'][indice] = media_idades
```
