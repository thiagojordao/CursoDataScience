```python
import pandas as pd
import numpy as np
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration',
       'actors_list']
file = 'movies.csv'
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns
```

```python
#Ordem descendente
filmes.title.sort_values(ascending=False).head(10)
```

```python
#Há outra forma de ordenar um DataFrame. Desta forma, o retorno é um DataFrame
filmes.sort_values('title').head(10)
```

```python
#Apenas ordenar não altera o Dataframe original
filmes.head()
```

```python
#Ordenando por genero e duração
filmes.sort_values(['genre','duration']).head(20)
```

```python
## Filtrando um DataFrame
```

```python
#Vamos criar uma lista para saber os filmes com maior duração do que 200 minutos

boleanos=[]

for item in filmes.duration:
    if item >=200:
        boleanos.append(True)
    else:
        boleanos.append(False)
```

```python
boleanos[:20] #é uma lista, não uma série
```

```python
type(boleanos)
```

```python
#Verificando o tamanho da lista
len(boleanos)
```

```python
filmes.shape
```

```python
#Precisamos converter a lista em uma SERIE para apllicar como filtro de Dataframe Pandas
```

```python
#Convertendo a lista em serie

duracao = pd.Series(boleanos)
duracao.head()
```

```python
type(duracao)
```

```python
#Podemos utilizar a notação de colchetes para indexar ou aplicar filtros de series boleanas
print(filmes['title'][:10])
```

```python
filmes[duracao]
```

```python
#Podemos ainda continuar a notação trazendo somente as colunas desejadas
filmes[duracao][['title','star_rating']]
```

```python
#Forma simplificada
A = filmes.duration >= 180
A.head(20)
```

```python
filmes[A].head()
```

```python
#Escrevendo por list Comprehension
filmes[filmes.duration >= 180].head()
```

```python
#Porém, por boa prática, utilizamos sempre o método LOC que nada mais é do que definir uma condição e filtrar por labels
#Veremos o ILOC em mais detalhes

filmes.loc[filmes.duration >= 200, 'genre']
```
