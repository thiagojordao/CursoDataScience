```python
import pandas as pd
import numpy as np
```

```python
colunas_filmes = ['star_rating', 'title', 'content_rating', 'genre', 'duration','actors_list']
file = 'movies.csv'
filmes = pd.read_csv(file, header=0, names=colunas_filmes)
filmes.columns
```

```python
#Filtrando com duração superior a 180 min

filmes[filmes.duration >= 180]
```

```python
#Operador AND
#Trazendo somente os filmes do gênero drama E com duração superior a 200 minutos

dramas_longos = filmes[(filmes.duration >=200) & (filmes.genre=="Drama")]
```

```python
dramas_longos
```

```python
#Operador Or
#Trazer somente os filmes fo gênero drama OU com duração superior a 200 minutos

filmes_longos = filmes[(filmes.duration >=200) | (filmes.genre=="Drama")]
```

```python
filmes_longos
```

```python
# Trazer os filmes dos gêneros crime, drama ou ação

filmes[(filmes.genre=="Crime") | (filmes.genre=="Drama") | (filmes.genre=="Ação")].head(10)
```

```python
# Utilizando o método ISIN
filmes[filmes.star_rating.isin([9.0,9.1,9.2,9.3])]
```

```python
filmes[filmes.genre.isin(["Crime","Drama","Ação"])].head()
```
